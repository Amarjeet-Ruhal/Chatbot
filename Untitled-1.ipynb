```python
import numpy as np
import pandas as pd
import json
```

```python
with open('intents.json') as fp:
    data=json.load(fp)
```

```python
inputs=[]
tag=[]
responses={}
for intent in data['intents']:
    responses[intent['tag']]=intent['responses']
    for inp in intent['patterns']:
        inputs.append(inp)
        tag.append(intent['tag'])
```

```python
data=pd.DataFrame({'inputs':inputs,'tags':tag})
data
```

```python
import string

data['inputs'] = [''.join(char for char in s if char not in string.punctuation).lower() for s in data['inputs']]
data
```

```python
unique_count = len(set(word for s in data['inputs'] for word in s.split()))
unique_count
```

```python
mx=0
for s in data['inputs']:
    mx=max(mx,len(s.split()))

mx
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.preprocessing import LabelEncoder
```

```python
texts = data['inputs']

tokenizer = Tokenizer(num_words=2000)
tokenizer.fit_on_texts(texts)
X = tokenizer.texts_to_sequences(texts)
X = pad_sequences(X)

# print(X[0]) 
```

```python
# Encode labels
y = data['tags']
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)
y
```

```python
voc=len(tokenizer.word_index)
voc
```

```python
output_len=label_encoder.classes_.shape[0]
output_len
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, Flatten

vocab_size = unique_count  # Size of vocabulary (based on tokenizer)
embedding_dim = 16  # Size of embedding vectors
max_length = 500     # Length of padded sequences

# Create the model
model = Sequential([
    Embedding(input_dim=vocab_size+1, output_dim=10, input_length=2000),
    LSTM(10, return_sequences=False),  # LSTM with 64 units
    Flatten(),      # Fully connected layer
    Dense(output_len, activation='softmax')     # Output layer (binary classification)
])

# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

```

```python
train=model.fit(X,y,epochs=200)
```

```python
import matplotlib.pyplot as plt
plt.plot(train.history['accuracy'],label='training data accuracy')
plt.plot(train.history['loss'],label='training data loss')
```

```python

import random

txt=input("You: ")
def output(txt):
    txt=[''.join(char for char in s if char not in string.punctuation).lower() for s in txt]
    tokenized = tokenizer.texts_to_sequences(txt)
    padded = pad_sequences(tokenized, maxlen=9)

    pred=model.predict(padded)

    pred=pred.argmax()
    pred=label_encoder.inverse_transform([pred])[0]
    print(pred)
    print("Bot :",random.choice(responses[pred]))

output(txt)
```
